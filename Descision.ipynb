```python
# Step 1: Import required libraries
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor, plot_tree
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import numpy as np

# Step 2: Load the Iris dataset
iris = load_iris()
X = iris.data  # All features
feature_names = iris.feature_names

# Step 3: Set up regression task
# We'll predict 'petal length (cm)' â€” that's feature at index 2
y_reg = X[:, 2]  # Target: petal length
X_reg = X[:, [0, 1, 3]]  # Features: sepal length, sepal width, petal width

# Step 4: Split into training and testing data
X_train, X_test, y_train, y_test = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)

# Step 5: Create and train the Decision Tree Regressor
dt_reg = DecisionTreeRegressor(max_depth=3, random_state=42)
dt_reg.fit(X_train, y_train)

# Step 6: Predict on the test set
y_pred = dt_reg.predict(X_test)

# Step 7: Evaluate the model
mse = mean_squared_error(y_test, y_pred)
print(f"ðŸ“Š Mean Squared Error (Petal Length Prediction): {mse:.2f}")

# -------------------- PLOT 1: Actual vs Predicted --------------------
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, color='blue', alpha=0.6, edgecolors='k', s=80)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--')
plt.xlabel("Actual Petal Length (cm)")
plt.ylabel("Predicted Petal Length (cm)")
plt.title("ðŸ“ˆ Actual vs Predicted Petal Length")
plt.grid(True)
plt.tight_layout()
plt.show()

# -------------------- PLOT 2: Decision Tree --------------------
plt.figure(figsize=(16, 8))
plot_tree(dt_reg,
          feature_names=["sepal length", "sepal width", "petal width"],
          filled=True,
          rounded=True,
          fontsize=10)
plt.title("ðŸŒ³ Decision Tree Structure for Regression")
plt.show()
```
